```python
import pandas as pd

df = pd.read_csv("wine-quality.csv")
df.head()
```

```python
df.isna().sum()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

df.plot.scatter(x="alcohol", y="quality", ax=axes[0])
df.plot.scatter(x="volatile acidity", y="quality", ax=axes[1])
df.plot.scatter(x="sulphates", y="quality", ax=axes[2])

plt.tight_layout()
plt.show()
```

```python
df.corr(numeric_only=True)["quality"].sort_values()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

X = df.drop(columns=["quality"])
y = df["quality"]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

model = ElasticNet(alpha=0.5, l1_ratio=0.5, random_state=42)
model.fit(X_train, y_train)

preds = model.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, preds))
mae = mean_absolute_error(y_test, preds)
r2 = r2_score(y_test, preds)

rmse, mae, r2
```

```python
results = X_test.copy()
results["actual_quality"] = y_test
results["predicted_quality"] = preds
results[["actual_quality", "predicted_quality"]].head(10)
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, preds, alpha=0.4)
plt.xlabel("Actual quality")
plt.ylabel("Predicted quality")
plt.title("Actual vs Predicted Wine Quality")
plt.show()
```

```python
results[["actual_quality", "predicted_quality"]].head(10).reset_index(drop=True)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)

rf_preds = rf.predict(X_test)

rf_rmse = np.sqrt(mean_squared_error(y_test, rf_preds))
rf_mae = mean_absolute_error(y_test, rf_preds)
rf_r2 = r2_score(y_test, rf_preds)

rf_rmse, rf_mae, rf_r2
```

```python
{
    "elasticnet": {"rmse": rmse, "mae": mae, "r2": r2},
    "random_forest": {"rmse": rf_rmse, "mae": rf_mae, "r2": rf_r2},
}
```

```python
comparison = pd.DataFrame(
    {
        "model": ["ElasticNet", "RandomForest"],
        "rmse": [rmse, rf_rmse],
        "mae": [mae, rf_mae],
        "r2": [r2, rf_r2],
    }
)

comparison
```
